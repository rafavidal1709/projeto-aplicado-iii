```python
import pdfplumber
import re
from google.colab import files

# Fazer o upload do arquivo PDF
uploaded = files.upload()

# Pegar o nome do arquivo original
for file_name in uploaded.keys():
    original_name = file_name
    break

# Carregar o PDF
with pdfplumber.open(original_name) as pdf:
    all_text = ""
    for page in pdf.pages:
        all_text += page.extract_text()

# Limpar o texto removendo múltiplos espaços, quebras de linha e caracteres especiais
cleaned_text = re.sub(r'\s+', ' ', all_text)  # Remove espaços extras e quebras de linha
cleaned_text = re.sub(r'[^\w\s,.!?]', '', cleaned_text)  # Remove caracteres especiais

# Criar o nome do arquivo de saída com a extensão .txt
output_name = original_name.replace('.pdf', '.txt')

# Salvar o texto limpo em um arquivo .txt
with open(output_name, 'w') as f:
    f.write(cleaned_text)

# Baixar o arquivo .txt gerado
files.download(output_name)
```
